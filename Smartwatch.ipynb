```python
nan_df = dailyActivity[dailyActivity.isna().any(axis=1)]
```

```python
dailyActivity = dailyActivity.dropna(how="any")
```

```python
dailyActivity["TotalSteps"] = pd.to_numeric(dailyActivity["TotalSteps"])
dailyActivity["TotalDistance"] = pd.to_numeric(dailyActivity["TotalDistance"])
dailyActivity["TrackerDistance"] = pd.to_numeric(dailyActivity["TrackerDistance"])
dailyActivity["LoggedActivitiesDistance"] = pd.to_numeric(dailyActivity["LoggedActivitiesDistance"])
dailyActivity["VeryActiveDistance"] = pd.to_numeric(dailyActivity["VeryActiveDistance"])
dailyActivity["ModeratelyActiveDistance"] = pd.to_numeric(dailyActivity["ModeratelyActiveDistance"])
dailyActivity["LightActiveDistance"] = pd.to_numeric(dailyActivity["LightActiveDistance"])
dailyActivity["SedentaryActiveDistance"] = pd.to_numeric(dailyActivity["SedentaryActiveDistance"])
#dailyActivity["VeryActiveMinutes"] = pd.to_numeric(dailyActivity["VeryActiveMinutes"])
#dailyActivity["LightlyActiveMinutes"] = pd.to_numeric(dailyActivity["LightlyActiveMinutes"])
#dailyActivity["SedentaryMinutes"] = pd.to_numeric(dailyActivity["SedentaryMinutes"])
dailyActivity["Calories"] = pd.to_numeric(dailyActivity["Calories"])
```

# Exploring the Relationship between Total Steps and Calories Burnt

```python
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
plt.scatter(dailyActivity["TotalSteps"], dailyActivity["Calories"])
plt.show()
correlation = round(dailyActivity["TotalSteps"].corr(dailyActivity["Calories"]),4)
print(correlation)
```
